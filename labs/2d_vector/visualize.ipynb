```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
import seaborn as sns 
plt.rcParams['text.usetex'] = True
plt.rcParams['font.size'] = 12

sns.set_style("whitegrid")
seed=21 # 8  14
np.random.seed(seed)
C1, C2 = 10, 100
range_x1 = [0 for i in  range(C1)]
range_x2 = [0 for i in  range(C2)]

M1, M2 = 0.1, 1.0
N1, N2 = 0.1, 0.1

U = [(np.random.random()-0.5)*M1 for i in range(C1)]
V = [(np.random.random()-0.5)*M2 for i in range(C1)]

U2 = [(np.random.random()-0.5)*N1 for i in range(C2)]
V2 = [(np.random.random()-0.5)*N2 for i in range(C2)]

S1 = [sum(U), sum(V)]
S2 = [sum(U2), sum(V2)]
S3 = [sum(U)+sum(U2), sum(V)+sum(V2)]

fig1, ax1 = plt.subplots(figsize=(4., 4.))
plt.axhline(0,  ls='--', c='gray')
plt.axvline(0,  ls='--', c='gray')
Q = ax1.quiver(range_x1, range_x1, U, V,  color='blue', angles='uv', alpha=0.4, width=0.010,
               headlength=5, scale=0.4, scale_units ='inches')# units='width',


Q = ax1.quiver(range_x2, range_x2, U2, V2,  color='red', angles='uv', alpha=0.4, width=0.010,
               headlength=5, scale=0.4, scale_units ='inches')# units='width',


Q = ax1.quiver([0], [0], [S1[0]], [S1[1]],  color='blue', angles='uv', width=0.018,
               headlength=5, scale=0.4, scale_units ='inches',)# units='width',

Q = ax1.quiver([0], [0], [S2[0]], [S2[1]],  color='red', angles='uv', width=0.018,
               headlength=5, scale=0.4, scale_units ='inches')# units='width',

Q = ax1.quiver([0], [0], [S3[0]], [S3[1]],  color='black', angles='uv', width=0.018,
               headlength=5, scale=0.4, scale_units ='inches')# units='width',


ax1.set_xlabel("Dim 1")
ax1.set_ylabel("Dim 2")
ax1.set_title(None)

plt.legend(['_nolegend_', '_nolegend_', '_nolegend_', '_nolegend_',  r'$\mathcal{H}^{+}, |\mathcal{H}|=10$', r'$\mathcal{L}^{+}, |\mathcal{L}|= 100$', r'$\mathcal{(H\cup C)}^{+}$'], loc='upper right', fontsize=10, )
plt.savefig("gradient_sum.pdf", bbox_inches='tight')
# qk = ax1.quiverkey(Q, 0.9, 0.9, 2, r'$2 \frac{m}{s}$', labelpos='E',
#                    coordinates='figure')
```
