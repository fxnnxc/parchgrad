```python
base_dir = '../../outputs'
import json 
import os 
import numpy as np 
from omegaconf import OmegaConf 

success_flags = []
for root, dirs, files in os.walk(base_dir):
    if 'config.yaml' in files:
        flags = OmegaConf.load(os.path.join(root, 'config.yaml'))
        if not flags.success:
            continue
        else:
            success_flags.append(flags)
```

```python
import pandas as pd 
tdf = pd.DataFrame(success_flags)
df = tdf.drop(columns=['alpha', 'bbox_path', 'data_path', 'device', 'exact_variance', 'gamma_infinity', 'success'])
tdf
```
